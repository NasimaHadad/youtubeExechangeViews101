## Section 2 : Type the keyword

```python
# Create Xpath code to Type keyword

# search = id="search" / name="search_query"
# search button = id="search-icon-legacy" / class="style-scope ytd-searchbox"
```

```python
# Type the keyword into the YouTube search engine
keyword = driver.find_element_by_xpath('//input[@aria-label="Search"]')

# Type in element
keyword.send_keys("حجز موعد فيزا اسبانيا 2021 ")
time.sleep(3)

```

```python
#Click search button
# Locate element
search_button = driver.find_element_by_id("search-icon-legacy")

# Click Element
search_button.click()
time.sleep(3)
```

#### --- Scroll down to the end of results page

```python
# Create a list from a single elemment
nomore_results = driver.find_elements_by_xpath('//yt-formatted-string[text()="No more results"]')

# print(nomore_results)

# Create a Conditional loop based on len(nomore_results) - the lenght of the list
while len(nomore_results) == 0:
    driver.execute_script("window.scrollBy(0,3000)")
    time.sleep(1)
    nomore_results = driver.find_elements_by_xpath('//yt-formatted-string[text()="No more results"]')

    
print("No More Results !!")    
    
```
